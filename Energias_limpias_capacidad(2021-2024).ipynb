### Energy capacity for 2021-2024

```python
cleanTech_capacity = data_capacity_2024.copy()
```

```python
cleanTech_capacity = cleanTech_capacity.iloc[7:20,:]
cleanTech_capacity
```

```python
cleanTech_capacity["+5%"] = [round((int(capacity) * 40) / 35,2) for capacity in list(cleanTech_capacity["2024"])] #This calculates the total energy capacity for 40%
def additionalEnergy(actualCapacity, expectedCapacity):
    values = []
    for index,actual in enumerate(actualCapacity):
        difference = expectedCapacity[index] - actual
        values.append(difference)
    return values
additionalEnergy = additionalEnergy(cleanTech_capacity["2024"].tolist(), cleanTech_capacity["+5%"].tolist())
cleanTech_capacity["Additional Capacity Energy (MW)"] = additionalEnergy
cleanTech_capacity
```

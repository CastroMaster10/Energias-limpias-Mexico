### Load the data

```python
data_generation_2024 = pd.read_csv("Generación de energía eléctrica 2021-2024 GigaWattsh.csv") #GigaWatts per hour
data_generation_2024
```

### Energy generation for 2021-2024

```python
cleanTech_generation = data_generation_2024.copy()
```

```python
cleanTech_generation = cleanTech_generation.iloc[7:19,:]
cleanTech_generation
```

```python
cleanTech_generation["+5%"] = [round((int(generation) * 40) / 35,2) for generation in list(cleanTech_generation["2024"])] #This calculates the total energy capacity for 40%
def additionalEnergy(actualGeneration, expectedGeneration):
    values = []
    for index,actual in enumerate(actualGeneration):
        difference = expectedGeneration[index] - actual
        values.append(difference)
    return values
additionalEnergy = additionalEnergy(cleanTech_generation["2024"].tolist(), cleanTech_generation["+5%"].tolist())
cleanTech_generation["Additional Generation (GW-h)"] = additionalEnergy
cleanTech_generation
```

```python
print("To reach the goal of having 40% of our energy from sustainable resources we need more plants to produce a total of " + str(round(cleanTech_generation["Additional Generation (GW-h)"][18], 2)) + " GigaWatts per hour.")
```
